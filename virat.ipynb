```python
import matplotlib.pyplot as plt  # type: ignore
import pandas as pd # type: ignore
import datetime as dt
import seaborn as sns
```

```python
df= pd.read_csv('data/Virat_Kolhi_Centuries.csv')
df.drop('Unnamed: 14',axis=1,inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
```

# 1. Centuries by Year

```python
plt.figure(figsize=(12, 6))
df['Year'].value_counts().sort_index().plot(kind='bar')
plt.title("Virat Kohli's Centuries by Year")
plt.xlabel("Year")
plt.ylabel("Number of Centuries")

plt.show()
```

# 2. Centuries by Format


```python
plt.figure(figsize=(8, 6))
df['Format'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Virat Kohli's Centuries by Format")
plt.show()
```

# 3. Centuries by Batting Order


```python

plt.figure(figsize=(10, 6))
df['Batting Order'].value_counts().plot(kind='bar')
plt.title("Virat Kohli's Centuries by Batting Order")
plt.xlabel("Batting Position")
plt.ylabel("Number of Centuries")
plt.show()
```

# 4. Strike Rate Distribution

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Strike Rate'], bins=20, kde=True)
plt.title("Distribution of Strike Rates in Kohli's Centuries")
plt.xlabel("Strike Rate")
plt.show()
```

# 5. Centuries Home vs Away

```python

plt.figure(figsize=(8, 6))
df['H/A'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Virat Kohli's Centuries: Home vs Away")

plt.show()
```

# 6. Top 10 Opponents

```python

plt.figure(figsize=(12, 6))
df['Against'].value_counts().nlargest(10).plot(kind='bar')
plt.title("Top 10 Opponents in Kohli's Centuries")
plt.xlabel("Opponent")
plt.ylabel("Number of Centuries")

plt.show()
```

# 7. Centuries as Captain vs Non-Captain

```python
plt.figure(figsize=(8, 6))
df['Captain'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Virat Kohli's Centuries: Captain vs Non-Captain")
plt.show()
```

# 8. Man of the Match Awards


```python

plt.figure(figsize=(8, 6))
df['Man of the Match'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Man of the Match Awards in Kohli's Centuries")
plt.show()

```
