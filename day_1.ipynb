```python
import matplotlib.pyplot as plt  # type: ignore
import pandas as pd # type: ignore
```

```python
data=pd.read_csv('data/baseball.csv')
# print(data.head())
print(data.keys())
print(data.describe())
height= data['Height']
weight =data['Weight']


```

```python
plt.plot(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height VS Weight")

plt.show()
```

```python
plt.hist(height,bins=5)
```

```python
plt.scatter(height,weight,alpha=0.5,c='red')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height and Weight Comparison")
plt.colorbar()

plt.show()
```
